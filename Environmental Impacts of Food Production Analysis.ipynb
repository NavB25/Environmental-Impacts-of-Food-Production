# Data Pre-Processing

## Missing Values

Let's first get a better understanding of the missing values in our data set.

```python
missing_values = df.isnull().sum()
percentage = round(100* df.isnull().sum()/len(df),2)
data_types = df.dtypes
missing_value_table=pd.concat([missing_values,percentage,data_types],axis=1)
missing_value_table = missing_value_table.rename(columns = {0:'Missing Values',1:'Percentage',2:'Data Types'})
missing_value_table
```

As shown in the table above, there are a number of columns with missing values. Later in our analysis, we will drop any NaN's specifically from the column of values we are analyzing.

## Data Reshaping

Before we conclude our pre-processing, we will create a new column called 'Food Type' and categorize our food products on whether they are animal-based or plant-based. 

```python
animal_based = []
plant_based = []

for food in df['Food product'][0:33]:
    plant_based.append(food)
for food in df['Food product'][33:]:
    animal_based.append(food)
    
df['Food Type'] = df['Food product'].apply(lambda x: 'Animal-Based' if x in animal_based else 'Plant-Based')
new_cols = ['Food product','Food Type','Land use change', 'Animal Feed', 'Farm', 'Processing',
       'Transport', 'Packging', 'Retail', 'Total_emissions',
       'Eutrophying emissions per 1000kcal (gPO₄eq per 1000kcal)',
       'Eutrophying emissions per kilogram (gPO₄eq per kilogram)',
       'Eutrophying emissions per 100g protein (gPO₄eq per 100 grams protein)',
       'Freshwater withdrawals per 1000kcal (liters per 1000kcal)',
       'Freshwater withdrawals per 100g protein (liters per 100g protein)',
       'Freshwater withdrawals per kilogram (liters per kilogram)',
       'Greenhouse gas emissions per 1000kcal (kgCO₂eq per 1000kcal)',
       'Greenhouse gas emissions per 100g protein (kgCO₂eq per 100g protein)',
       'Land use per 1000kcal (m² per 1000kcal)',
       'Land use per kilogram (m² per kilogram)',
       'Land use per 100g protein (m² per 100g protein)',
       'Scarcity-weighted water use per kilogram (liters per kilogram)',
       'Scarcity-weighted water use per 100g protein (liters per 100g protein)',
       'Scarcity-weighted water use per 1000kcal (liters per 1000 kilocalories)']
df = df[new_cols]
df.head()
```

# Exploratory Analysis

Excellent! We are now ready to begin our exploratory analysis of the data. Let's start by understanding what type of foods create the most greenhouse gas (GHG) emissions per kg of product. 

### Greenhouse Gas Emissions per Kilogram (measured in kgC02 equivalents per kg)

```python
food_type = df.groupby(['Food Type'])['Total_emissions'].sum().reset_index()
explode = (0.1, 0)
colors = ['darksalmon','forestgreen']
plt.figure(figsize=(5,5))
plt.pie(food_type['Total_emissions'], labels=food_type['Food Type'],
    autopct='%.0f%%',startangle=90,explode=explode,shadow=True,textprops={'fontsize': 15},colors=colors);
plt.title('Total Greenhouse Gas Emissions by Food Type\nAnimal-Based vs Plant-Based',fontsize=15);
```

Animal-based food products emit more GHG emissions than plant-based food products. Knowing this, let's shift our focus to the supply chain stages and pinpoint at what stage we are seeing the greatest number of emissions being emitted and which food type is the culprit. 

```python
supply_chain = df.iloc[:,1:9].groupby('Food Type').sum().reset_index()

x = supply_chain.columns[1:].values

X_axis = np.arange(len(x))
width = 0.4

y1 = supply_chain.iloc[0,1:8].values
y2 = supply_chain.iloc[1,1:8].values

fig, ax =  plt.subplots(figsize = (12, 9))
ax.bar(X_axis - 0.2,y1,label = 'Animal Based',color = 'darksalmon',width=width)
ax.bar(X_axis + 0.2,y2, label = 'Plant Based', color = 'forestgreen',width=width)

for bar in ax.patches:
    ax.annotate(s = str(round(bar.get_height(),2)),
                  xy = (bar.get_x() + bar.get_width() / 2, bar.get_height()+1),
                  ha='center', 
                  va='center',
                  size=12,
                  xytext=(0, 6),
                  textcoords='offset points')



plt.title('Total Greenhouse Gas Emissions by Supply Chain Stage & Food Type \n',fontsize=15)
plt.ylabel('kgCO2eq/kg',fontsize=15)
plt.xticks(X_axis, x)
plt.xlabel('\nSupply Chain Stage',fontsize=15)


plt.legend()
plt.tight_layout()
```

Animal-based products at the 'Farm' stage of the supply chain release the greatest amount emissions in comparison to all the other stages. We can go even deeper in our analysis and figure out which specific food products are the largest emitters of GHG emissions. 

```python
plt.figure()
plt.style.use('ggplot')
plots = df.sort_values('Total_emissions').plot(kind='barh',x='Food product',y='Total_emissions',color='seagreen',figsize=(20,15),legend=False,fontsize=15,edgecolor='black')

for i in plots.patches:
    plots.text(i.get_width()+.5, i.get_y(), \
            str(round(i.get_width(),2)), fontsize=15,
            color='black')

plt.title('Total Greenhouse Gas Emissions per Kilogram of Food Product \n',fontsize=18)
plt.ylabel('')
plt.xlabel('\nkgCO2eq/kg',fontsize=18)
plt.tight_layout()
```

We should make note of the number of animal-based products that appear at the top of our plot. Our observation earlier regarding emissions caused by animal-based products at the 'Farm' stage of the supply chain is now making a lot more sense. Methane produced by livestock on farms contributes to large amounts of GHG emissions released into the air.

### Eutrophying Emissions (measured in gPO4 equivalents per kg)

Eutrophication is the pollution of water bodies and ecosystems with excess nutrients. Agricultural production systems release nitrogen and nutrients which can be a major environmental problem. Eutrophying emissions are measured in grams of phosphate equivalents - P04eq.

```python
df2 = df[['Food product','Eutrophying emissions per kilogram (gPO₄eq per kilogram)']].dropna()
plt.figure()
plots = df2.sort_values('Eutrophying emissions per kilogram (gPO₄eq per kilogram)').plot(kind='barh',x='Food product',y='Eutrophying emissions per kilogram (gPO₄eq per kilogram)',color='r',figsize=(20,15),legend=False,fontsize=15,edgecolor='black')

for i in plots.patches:
    plots.text(i.get_width()+2, i.get_y(), \
            str(round(i.get_width(),2)), fontsize=15,
            color='black')

plt.title('Eutrophying Emissions per Kilogram of Food Product \n',fontsize=18)
plt.ylabel('')
plt.xlabel('\ngPO4eq/kg)',fontsize=18)
plt.tight_layout()
```

### Freshwater Withdrawals (measured in liters per kg)

```python
df3 = df[['Food product','Freshwater withdrawals per kilogram (liters per kilogram)']].dropna()
plt.figure()
plt.style.use('ggplot')
plots = df3.sort_values('Freshwater withdrawals per kilogram (liters per kilogram)').plot(kind='barh',x='Food product',y='Freshwater withdrawals per kilogram (liters per kilogram)',color='dodgerblue',figsize=(20,15),legend=False,fontsize=15,edgecolor='black');

for i in plots.patches:
    plots.text(i.get_width()+20, i.get_y(), \
            str(int(round(i.get_width()))), fontsize=15,
            color='black')

plt.title('Freshwater Withdrawals per Kilogram of Food Product\n',fontsize=18);
plt.ylabel('');
plt.xlabel('\nL/kg)',fontsize=18)
plt.tight_layout()
```

### Land Use (measured in m² per kg)

```python
df4 = df[['Food product','Land use per kilogram (m² per kilogram)']].dropna()
plt.figure()
plt.style.use('ggplot')
plots = df4.sort_values('Land use per kilogram (m² per kilogram)').plot(kind='barh',x='Food product',y='Land use per kilogram (m² per kilogram)',color='darkgoldenrod',figsize=(20,15),legend=False,fontsize=15,edgecolor='black')

for i in plots.patches:
    plots.text(i.get_width()+2, i.get_y(), \
            str(round(i.get_width(),2)), fontsize=15,
            color='black')

plt.title('Land Use per Kilogram of Food Product\n',fontsize=18)
plt.ylabel('',fontsize=18)
plt.xlabel('\nm²/kg',fontsize=18)
plt.tight_layout()
```

### Scarcity-Weighted Water Use (measured in liters per kg)

```python
df5 = df[['Food product','Scarcity-weighted water use per kilogram (liters per kilogram)']].dropna()

plt.figure()
plt.style.use('ggplot')
plots = df5.sort_values('Scarcity-weighted water use per kilogram (liters per kilogram)').plot(kind='barh',x='Food product',y='Scarcity-weighted water use per kilogram (liters per kilogram)',color='mediumpurple',figsize=(20,15),legend=False,fontsize=15,edgecolor='black')

for i in plots.patches:
    plots.text(i.get_width()+500, i.get_y(), \
            str(int(round(i.get_width()))), fontsize=15,
            color='black')

plt.title('Scarcity-Weighted Water Use per Kilogram of Food Product \n',fontsize=18)
plt.ylabel('',fontsize=18)
plt.xlabel('\nL/kg)',fontsize=18)
plt.tight_layout()
```

Let's compare each of the columns we've analyzed above with the use of pie plots! This will give us a better idea of the proportions of each food type that makes up the total for each column.

```python
eutrophying = df[['Food Type','Eutrophying emissions per kilogram (gPO₄eq per kilogram)']].dropna().groupby(['Food Type'])['Eutrophying emissions per kilogram (gPO₄eq per kilogram)'].sum().reset_index()
freshwater = df[['Food Type','Freshwater withdrawals per kilogram (liters per kilogram)']].dropna().groupby(['Food Type'])['Freshwater withdrawals per kilogram (liters per kilogram)'].sum().reset_index()
land = df[['Food Type','Land use per kilogram (m² per kilogram)']].dropna().groupby(['Food Type'])['Land use per kilogram (m² per kilogram)'].sum().reset_index()
scarcity = df[['Food Type','Scarcity-weighted water use per kilogram (liters per kilogram)']].dropna().groupby(['Food Type'])['Scarcity-weighted water use per kilogram (liters per kilogram)'].sum().reset_index()

fig, axs = plt.subplots(2, 2,figsize =(12,12))

explode = (0.1, 0)
colors = ['darksalmon','forestgreen']

axs[0, 0].pie(eutrophying['Eutrophying emissions per kilogram (gPO₄eq per kilogram)'], labels=eutrophying['Food Type'],
    autopct='%.0f%%',startangle=90,explode=explode,shadow=True,textprops={'fontsize': 15},colors=colors);
axs[0, 0].set_title('Eutrophying Emissions by Food Type',fontsize=15)


axs[0, 1].pie(freshwater['Freshwater withdrawals per kilogram (liters per kilogram)'], labels=freshwater['Food Type'],
    autopct='%.0f%%',startangle=90,explode=explode,shadow=True,textprops={'fontsize': 15},colors=colors);
axs[0, 1].set_title('Freshwater Withdrawals by Food Type',fontsize=15)


axs[1, 0].pie(land['Land use per kilogram (m² per kilogram)'], labels=land['Food Type'],
    autopct='%.0f%%',startangle=90,explode=explode,shadow=True,textprops={'fontsize': 15},colors=colors);
axs[1, 0].set_title('Land Use by Food Type',fontsize=15)


axs[1, 1].pie(scarcity['Scarcity-weighted water use per kilogram (liters per kilogram)'], labels=scarcity['Food Type'],
    autopct='%.0f%%',startangle=90,explode=explode,shadow=True,textprops={'fontsize': 15},colors=colors);
axs[1, 1].set_title('Scarcity-Weighted Water Use by Food Type',fontsize=15)

plt.tight_layout()
```

A common theme in both the pie and horizontal bar plots above is how often we see animal-based products making up a greater proportion of the total for each column - with the exception of Scarcity-Weighted Water Use. It is only reasonable to conclude that overall, based on our analysis of the dataset, the production of animal-based food products has a greater environmental impact compared to plant-based foods.
